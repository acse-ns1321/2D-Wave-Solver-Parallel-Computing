## SERIAL CODE -NEUMANN BOUNDARIES
The following Script runs and animates the serial code for the 2D wave solver 

```python
# Setup  figure
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

def animation_function(i):
    # Format the file string
    output_file =  "..\output/serial/output_{}.dat".format(i)
    
    # Load data from file
    data = np.loadtxt(output_file, dtype=float)
    ax.clear()
    # Plot the data
    ax.imshow(data, cmap="bone")

# Initialize the animation object
ani = matplotlib.animation.FuncAnimation(fig, animation_function, frames=750, repeat=True)
# plt.show()

# # Initialize the writer function
# writergif = animation.PillowWriter(fps=30) 
# # # Save the animation using the writer
ani.save("..\postprocessing/animate_serial.gif", writer=writergif)
# ani.save('..\postprocessing/animate_serial.mp4', fps=30)
```

## PARALLEL CODE - NEUMANN BOUNDARIES

```python
#-------------------------------------------------------------------------------------------
# OUTPUTS MUST BE INPUTTED BY THE USER DEPENDING ON OUTPUT PARAMETERS OBTAINED AFTER SIMULATION IS RUN
#-------------------------------------------------------------------------------------------
num_outputs = 1500
sub_rows = 2
sub_cols = 4
processors = 8
#-----------------------------USER INPUTS COMPLETE----------------------------------------

def recombine_data(iteration):
    data = []
    for i in np.arange(sub_rows):
        data_row = []
        for j in np.arange(sub_cols):
            process_data = np.loadtxt("..\output/parallel/neumann/{}_processors/output_{}_id_{}.dat".format(processors, iteration, i*sub_cols + j), dtype=float)
            data_row.append(process_data)
            
        combined_row = np.hstack(data_row)
        data.append(combined_row)

    return np.vstack(data)


fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

def animation_function(i):
    data = recombine_data(i)
    ax.clear()
    # Plot the data
    ax.imshow(data,cmap="bone" )
    
# Create animation and write to current directory
ani = matplotlib.animation.FuncAnimation(fig, animation_function, frames=num_outputs, interval=500)
writergif = animation.PillowWriter(fps=24) 
ani.save("..\postprocessing/animations/animate_parallel_neumann.gif", writer=writergif)
```
